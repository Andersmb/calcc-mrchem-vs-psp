```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import json
from traceback import format_exc
import numpy as np

from functions import getSCFCycleData, getMolname, getThresholds, getMult

# Get completed molecule indeces
project = 'row3'
data = f'{project}_total_energies.csv'
df = pd.read_csv(data)
molecules = df.loc[df.Completed]['Index'].unique().tolist()
functionals = df.Functional.unique().tolist()
functionals = ['pbe0']
```

```python
for mol in molecules:
    fig, axes = plt.subplots(ncols=len(functionals), dpi=100, figsize=(8, 4), sharey=True)
    if not isinstance(axes, np.ndarray):
        axes = [axes]
    
    for func, ax in zip(functionals, axes):
        try:
            fname = f'calcs_{project}/{mol}_{func}_.json'
            en, mo, up = zip(*getSCFCycleData(fname))
            ax.plot(mo, marker='.', color='blue', label='MO resiudal')
            ax.plot(up, marker='.', color='crimson', label='Energy update')
            
            ax2 = ax.twinx()
            ax2.plot(en, marker='.', color='black')
            ax2.ticklabel_format(useOffset=False)
            
            mo_thrs, up_thrs = getThresholds(fname)
            ax.axhline(mo_thrs, ls='--', color='blue')
            ax.axhline(up_thrs, ls='--', color='crimson')
            
            ax.set_title(f'{mol}_{getMolname(mol, project)}_{func}')
            ax.grid(ls=':', lw=0.5)
            ax.set_yscale('log')

            ax.legend(fontsize='small', loc='upper right')
            
            plt.tight_layout(h_pad=1)
            plt.savefig(f'convergence_plots/{mol}_{getMolname(mol, project)}_{func}.png', facecolor='white')
            
        except Exception as e:
            print(f'Skipped', mol, getMolname(mol, project), func)
            print(format_exc())
            print('')
            continue
```
